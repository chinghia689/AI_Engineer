```python
from sklearn.mixture import GaussianMixture
from sklearn.preprocessing import StandardScaler
X_all=np.r_[df_train,df_test]
ss=StandardScaler()
X_all=ss.fit_transform(X_all)
```

```python
from sklearn.decomposition import PCA
pca=PCA(n_components=0.99,whiten=True,random_state=42)
X_all=pca.fit_transform(X_all)
```

```python
gmm=GaussianMixture(n_components=4,covariance_type='full',random_state=42)
```

```python
gmm.fit(X_all)
```

```python
X_all=gmm.predict_proba(X_all)
```

```python
gmm_Train=X_all[:len(df_train)]
gmm_Test=X_all[len(df_train):]
```

```python
gmm_Train.shape
```

```python
df_train_lable=df_train_lable.values.reshape(-1)
```

```python
df_train_lable.shape
```

```python
X_train_gmm,X_test_gmm,y_trainbest,y_testbest=train_test_split(gmm_Train,df_train_lable,test_size=0.2,random_state=42)
```

```python
from sklearn.model_selection import GridSearchCV
rfc=RandomForestClassifier(random_state=42)
param_grid = {
    'n_estimators': [100, 200, 400], 
    'max_depth': [10, 20, 40, None], 
    'min_samples_leaf': [1, 2, 4],   
    'bootstrap': [True, False]
}
grid_search_rfc=GridSearchCV(
    estimator=rfc, 
    param_grid=param_grid, 
    cv=10, 
    scoring='accuracy', 
    n_jobs=-1   
).fit(X_train_gmm,y_trainbest)
```

```python
best_model = grid_search_rfc.best_estimator_
```

```python
score_best=grid_search_rfc.score(X_test_gmm,y_testbest)
```
