```python
import numpy as np
import tensorflow
from keras import models, preprocessing
import matplotlib.pyplot as plt
import cv2
```

```python
model=models.load_model('real_fake.keras')
```

```python
df=cv2.imread("/home/chinghia/AI_Engineer/Data/unnamed.jpg")
```

```python
df.shape
```

```python
df=cv2.cvtColor(df,cv2.COLOR_BGR2RGB)
```

```python
df=cv2.resize(df,(32,32))
```

```python
df.shape
```

```python
plt.imshow(df)
plt.show()
```

```python
df=df.reshape(-1,32,32,3)
```

```python
y_pred=model.predict(df)
```

```python
def label_name_pred(test):
    if test > 0.5:
        return('Real')
    else:
        return('Fake')


plt.imshow(df.reshape(32,32,3))
plt.title(label_name_pred(np.argmax(y_pred)))
plt.show()
```
